```python
class_counts = df['CLASS_LABEL'].value_counts()
print(class_counts)
```

```python
plt.figure(figsize=(8, 6)) #Tạo một hình vẽ mới với kích thước 8x6 inches
class_counts.plot(kind='bar', color='yellow') #Vẽ biểu đồ cột của số lượng mẫu trong mỗi nhóm (giá trị) của cột "CLASS_LABEL"
plt.title('Class Label NTU_Phishing') #đặt tên cho biểu đồ
plt.xlabel('Class Label') #đặt nhãn cho trục x
plt.ylabel('Count') #đặt nhãn cho trục y
```

```python
X = df.drop(columns=['CLASS_LABEL', 'id', 'HttpsInHostname','PathLength','SubdomainLevelRT',
                     'RightClickDisabled','UrlLength','PopUpWindow','PctExtResourceUrlsRT',
                     'NumHash','DoubleSlashInPath','AtSymbol','FakeLinkInStatusBar',
                     'UrlLengthRT','HostnameLength','EmbeddedBrandName','IpAddress',
                     'DomainInSubdomains','TildeSymbol','RandomString']) #Tạo một DataFrame mới là X để loại bỏ các cột không cần thiết
#Tạo một y chứa cột 'CLASS_LABEL', là biến mục tiêu cần dự đoán.
y = df['CLASS_LABEL']
 #Tạo một đối tượng StandardScaler để chuẩn hóa dữ liệu.
scaler = StandardScaler()
#Chuẩn hóa dữ liệu đầu vào X (đảm bảo rằng các đặc trưng sẽ có giá trị trung bình gần bằng 0 và độ lệch chuẩn gần bằng 1)
X = scaler.fit_transform(X)
```

RandomForest

```python
model = load("drive/MyDrive/archive/RF_model_2.joblib")
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
predictions = model.predict(X)
print(predictions)
# Dự đoán và đánh giá mô hình
accuracy = accuracy_score(y, predictions)
print(f'Accuracy: {accuracy * 100:.2f}%')
print(classification_report(y, predictions))
print("Confusion Matrix:")
conf_matrix = confusion_matrix(y, predictions)
sns.heatmap(conf_matrix, annot=True, cmap='Purples')
```

DecisionTree

```python
model = load("drive/MyDrive/archive/DT_model_2.joblib")
```

```python
from sklearn.metrics import accuracy_score,classification_report
X_testDT = model.predict(X)
print(X_testDT)
score = accuracy_score(y, X_testDT)
print(f'Accuracy: {score * 100:.2f}%')
print(classification_report(y, X_testDT))
conf_matrix = confusion_matrix(y, X_testDT)
sns.heatmap(conf_matrix, annot=True, cmap='Oranges')
```

Xgboots

```python
import xgboost as xgb
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
model = load("drive/MyDrive/archive/Xgb_model_2.joblib")
X_test_xgb = model.predict(X)
print(X_test_xgb)
accuracyxgb = accuracy_score(y, X_test_xgb)
print(f'Accuracy: {accuracyxgb * 100:.2f}%')
report1 = classification_report(y, X_test_xgb)
print("Classification Report:")
print(report1)
print("Confusion Matrix:")
conf_matrix = confusion_matrix(y, X_test_xgb)
sns.heatmap(conf_matrix, annot=True, cmap='Blues')
```

```python
accuracies = [accuracy * 100, score * 100, accuracyxgb * 100]
models = ['Random Forest', 'Decision Tree', 'XGBoost']

print(f'Accuracy (Random Forest): {accuracies[0]:.2f}%')
print(f'Accuracy (Decision Tree): {accuracies[1]:.2f}%')
print(f'Accuracy (XGBoost): {accuracies[2]:.2f}%')

plt.figure(figsize=(10, 6))
plt.bar(models, accuracies, color=['yellow', 'green', 'red'])
plt.xlabel('Models')
plt.ylabel('Accuracy (%)')
plt.title('Comparison of Model Accuracies')
plt.ylim(0, 100)
plt.show()
```
